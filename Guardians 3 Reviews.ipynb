```python
import pandas as pd
import csv
import time
import json
import requests
import warnings
import pandas as pd
warnings.filterwarnings("ignore")
```

```python
from selenium import webdriver

from selenium.webdriver.common.by import By

from selenium.webdriver.chrome.service import Service

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.support import expected_conditions as EC

import codecs

import re

from webdriver_manager.chrome import ChromeDriverManager
```

```python
def get_reviews():
    result = {}
    count = 0
    try:
        driver = webdriver.Chrome('chromedriver.exe')
        val = driver.get("https://www.rottentomatoes.com/m/guardians_of_the_galaxy_vol_3")
    except Exception as e:
        print(e)
    try:
        reviews = driver.find_element(By.XPATH, '//*[@id="critics-reviews"]/h2/a/em')
        reviews.click()
    except Exception as e:
        print(e)
    try:
        for j in range(10):
            time.sleep(10)
            reviewer_name = driver.find_elements(By.CLASS_NAME, 'display-name')
            review_text = driver.find_elements(By.CLASS_NAME, 'review-text')
            for i in range(len(reviewer_name)):
                result[reviewer_name[i].text]=review_text[i].text
            nextbutton = driver.find_elements(By.CLASS_NAME, 'next')
            nextbutton[0].click()
            time.sleep(10)
    except Exception as e:
        print(e)
    finally:
        driver.quit()
    return result
    

   
```

```python
result = get_reviews()
```

```python
result
```

```python
names,reviews = [],[]
for keys in result.keys():
    names.append(keys)
for values in result.values():
    reviews.append(values)
```

```python
data = {'Name': names,
        'Review': reviews}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('reviews.csv', index=False)
```
